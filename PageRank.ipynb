```python
M = np.array(P)
M[:, 40]
```

```python
M[34, 34] = 1
M[35, 35] = 1
M[36, 36] = 1
M[38, 38] = 1
M[34, 40] = 1
M[35, 40] = 1
M[36, 40] = 1
M[38, 40] = 1
```

```python
M = M.astype(np.float64)
for i in range(0, 40):
    for j in range(0, 41):
        M[i, j] = M[i, j]/M[i, 40]
#np.set_printoptions(threshold=np.nan, linewidth=100)
```

```python
M = np.delete(M, -1, axis=1)
```

```python
v = np.ones((40, 1), dtype = float)
v = v/40
```

```python
e = np.ones((1, 40), dtype = float)
```

```python
D = e*v
```

```python
a = 0.85
```

```python
M = a*M + (1-a)*D
```

```python
M = M.T
```

```python
import timeit
start=timeit.default_timer()
```

```python
def power_iteration (A, num_simulations):
    b_k = np.random.rand(A.shape[0])
    for _ in range(num_simulations):
        b_k1 = np.dot(A, b_k)
        b_k1_norm = np.linalg.norm(b_k1)
        b_k = b_k1 / b_k1_norm
    return b_k
z = np.around(power_iteration(M, 100), decimals=3)
print(z)
```

```python
end=timeit.default_timer()
print('Running time: %s Seconds'%(end-start))
```
